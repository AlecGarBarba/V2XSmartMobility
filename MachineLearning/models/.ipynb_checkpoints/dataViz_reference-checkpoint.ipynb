## Creating a healthy dataset, without any anomaly on the road

```python
parkdf= pd.read_csv('rec_csv_file_park_1.csv',usecols=['Ax','Ay','Az'])
df_plain_1= pd.read_csv('rec_csv_file_plain_street_1.csv',usecols=['Ax','Ay','Az'])  #until 410
df_plain_1= df_plain_1.head(410)
df_plain_1_1 = df_plain_1.head(300)
df_plain_1_2 = df_plain_1.tail(110)
df_plain_2= pd.read_csv('rec_csv_file_plain_street_2.csv',usecols=['Ax','Ay','Az'])  #completo
df_plain_3= pd.read_csv('rec_csv_file_plain_street_3.csv',usecols=['Ax','Ay','Az'])  #completo. Es el más ruidoso
df_plain_3= df_plain_3.tail(240)
df_plain_4= pd.read_csv('rec_csv_file_plain_street_3_2.csv',usecols=['Ax','Ay','Az']) 
```

```python
frames = []
frames = [parkdf, df_plain_1,df_plain_3,df_plain_4,df_plain_2, parkdf[:30],df_plain_4,df_plain_2[50:],df_plain_3[:100],parkdf[:20],df_plain_2,df_plain_4,df_plain_3[:150],df_plain_1]
```

## Creating replications of each healthy frame

```python
count = 0
for frame in frames:
    while frame.shape[0] < 300:
        frame = pd.concat([frame,frame])
    frame = frame.head(300)
    frames[count] = frame
    count+=1
```

```python
healthy_df = pd.concat(frames).reset_index(drop=False)
healthy_df.to_csv('anomaly_train_dataset.csv')
healthy_df.shape
```

```python
def treatFrames(frames):
    count = 0
    for  frame in frames:
        while frame.shape[0] < 300:
            frame = pd.concat([frame,frame])
        frame = frame.head(300)
        frames[count] = frame
        count+=1
    return frames
```
