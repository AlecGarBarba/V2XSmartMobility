```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_csv("cdmxAnalysis_.csv");
```

```python
#df.rename(columns = {'high Development': 'High Development', 'lowDevelopment': 'Low Development'}, inplace = True)
df.head(20)
```

```python
df["Average Income"] = df["Average Income"].astype(str).astype(float)
df.dtypes
```

```python
corr =df.to_numeric().corr()
```

```python
corr
```

```python
# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
f, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr-1, dtype=bool))
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0, annot = True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
