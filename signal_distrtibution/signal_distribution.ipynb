```python
import numpy as np  # scientific computation
import pandas as pd # primary data structure library
import matplotlib.pyplot as plt #primary plotting stricture

from datetime import datetime, timedelta
#Data visualization
from matplotlib import pyplot
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
def createDataFrameFromtxt(txt, packets_sent):
    arr = [];
    count = 0;
    with open(txt) as f:
        lines = f.readlines()
        for line in lines: 
            line =line.replace("<", "=")
            if(line[0] == "R"):
                #print(line)
                [_,li] = line.split("tiempo=")
                [ms,_] = li.split("ms") 
                arr.append(int(ms))
            else:
                count+=1;
    df = pd.DataFrame(data=arr, columns=["time"]).sort_values(by=["time"])
    res = df.groupby('time').size().to_frame('size').reset_index(drop=False)
    res['cumsum'] =res['size'].cumsum()
    res['Cumulative sum [%]'] = res['cumsum']/(packets_sent-count)*100
    res['time [ms]']= res['time']
    return res
```

```python
df70_100 = createDataFrameFromtxt('70m_100b.txt', 50);
df70_1000 = createDataFrameFromtxt('70m_1000b.txt', 50);
df80_100 = createDataFrameFromtxt('80m_100b.txt', 50);
df80_1000 = createDataFrameFromtxt('80m_1000b.txt', 50);

df90_100 = createDataFrameFromtxt('90m_100b.txt', 50);
df90_1000 = createDataFrameFromtxt('90m_1000b.txt', 50);

df100_100 = createDataFrameFromtxt('100m_100b.txt', 50);
df100_1000 = createDataFrameFromtxt('100m_1000b.txt', 50);


df110_100 = createDataFrameFromtxt('110m_100b.txt', 25);
df110_1000 = createDataFrameFromtxt('110m_1000b.txt',25);


df120_100 = createDataFrameFromtxt('120m_100b.txt',25);
df120_1000 = createDataFrameFromtxt('120m_1000b.txt',25);

```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))
plt.rcParams.update({'font.size': 22})




fig.suptitle('Cumulative distribution for 70 m distance')

sns.plotting_context(font_scale=5)
sns.lineplot(ax=axes[0],data=df70_100, x="time [ms]", y="Cumulative sum [%]").set_title('70 m 100 b')
             

axes[0].set_title("100 bytes")
sns.lineplot(ax=axes[1],data=df70_1000, x="time [ms]", y="Cumulative sum [%]").set_title('70 m 100 b')  
axes[1].set_title("1000 bytes") 
             
             
plt.rc('font', size=22)          # controls default text sizes
plt.rc('axes', titlesize=22)     # fontsize of the axes title
plt.rc('axes', labelsize=22)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=22)    # fontsize of the tick labels
plt.rc('ytick', labelsize=22)    # fontsize of the tick labels
plt.rc('legend', fontsize=22)    # legend fontsize
plt.rc('figure', titlesize=22)

plt.savefig('cdist_70.png')
```
