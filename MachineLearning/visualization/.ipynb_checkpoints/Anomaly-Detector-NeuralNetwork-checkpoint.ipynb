```python
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        sys.stdout.write(str(logs.get('loss'))+str(', '))
        sys.stdout.flush()
        self.losses.append(logs.get('loss'))
        
lr = LossHistory()
```

```python
def scaleData(data):
    # normalize features
    scaler = MinMaxScaler(feature_range=(-2, 2))
    return scaler.fit_transform(data)
```

```python
def reshape(data):
    data_scaled = scaleData(data)
    data_scaled = data_scaled.T
    return data_scaled.reshape(3, 300)
```

```python
new_dataset = pd.read_csv('anomaly_train_dataset.csv',  usecols=['Ax','Ay','Az'])
data_broken = pd.read_csv('anomaly_train_dataset_broken.csv',  usecols=['Ax','Ay','Az'])
new_dataset.head()
```

## Segment into 300 data points each and every thingy

```python
label_broken = np.repeat(1,3)
label_broken.shape = (3,1)
label_healthy = np.repeat(0,3)
label_healthy.shape = (3,1)
```

```python
# healthy 4200/300 = 12. broken 9900/300 = 33
np_healthy_fft_array = [] 
count = 0
length = 300
loop = True
while loop:
    if(count*length >=4200):
        loop = False
        break
    np_healthy_fft_array.append(reshape(data_healthy_fft[count*300 : count*300+length]))
    count +=1
```

```python
#np_healthy_fft_array[12]
```

```python
np_broken_fft_array = [] 
count = 0
length = 300
loop = True
while loop:
    if(count*length >=9900):
        loop = False
        break
    np_broken_fft_array.append(reshape(data_broken_fft[count*300 : count*300+length]))
    count +=1
```

```python

np.shape(np_broken_fft_array)
```

```python


train_healthy = np.hstack((np_healthy_fft_array,label_healthy))
train_broken = np.hstack((np_broken_fft_array,label_broken))
train_both = np.vstack((train_healthy,train_broken))
```

```python
pd.DataFrame(train_both)
```

```python
features = train_both[:,:2510]
labels = train_both[:,2510:]
```

## Training started

```python
dim = 300
samples = 3
```

```python
number_of_neurons_layer1 = 10 
number_of_neurons_layer2 = 5
number_of_neurons_layer3 = 1
number_of_epochs = 100
```

```python

sgd = optimizers.SGD(lr=0.01, clipnorm=1.)

model = Sequential()
model.add(Dense(number_of_neurons_layer1,input_shape=(dim, ), activation='relu'))
model.add(Dense(number_of_neurons_layer2, activation='relu'))
model.add(Dense(number_of_neurons_layer3, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=sgd)

def train(data,label):
    model.fit(data, label, epochs=number_of_epochs, batch_size=72, validation_data=(data, label), verbose=0, shuffle=True,callbacks=[lr])

def score(data):
    return model.predict(data)
```

```python
train(features,labels)
```

```python
fig, ax = plt.subplots(num=None, figsize=(14, 6), dpi=80, facecolor='w', edgecolor='k')
size = len(lr.losses)
ax.plot(range(0,size), lr.losses, '-', color='blue', animated = True, linewidth=1)
```
